### 2.6.3 Categorical Features<a id='2.6.3_Categorical_Features'></a>

So far you've examined only the numeric features. Now you inspect categorical ones such as description. These are discrete entities. 'North Park Hill' is a name. Although names can be sorted alphabetically, it makes no sense to take the average of 'North Park Hill' and 'Hale'. Similarly, 'North Park Hill' is after 'Hale' only lexicographically; it is neither 'greater than' nor 'less than' 'Hale'. As such, they tend to require different handling than strictly numeric quantities. Note, a feature _can_ be numeric but also categorical. Although you digress a little to make a point, month numbers are also, strictly speaking, categorical features. Yes, when a month is represented by its number (1 for January, 2 for Februrary etc.) it provides a convenient way to graph trends over a year. And, arguably, there is some logical interpretation of the average of 1 and 3 (January and March) being 2 (February). However, clearly December of one years precedes January of the next and yet 12 as a number is not less than 1. The numeric quantities in the section above are truly numeric; they are the number of feet in the drop, or acres or years open or the amount of snowfall etc.

Fortunately, we don't have any missing value in categorical features except the description column. Right now, I'm trying to select the kinds of homes from the description column.

A fairly new groupby behaviour is [named aggregation](https://pandas-docs.github.io/pandas-docs-travis/whatsnew/v0.25.0.html). This allows us to clearly perform the aggregations you want whilst also creating informative output column names.

```python
airbnb_data['neighbourhood_cleansed'].value_counts()
```

```python
#Add named aggregations for the sum of 'daysOpenLastYear', 'TerrainParks', and 'NightSkiing_ac'
#call them 'state_total_days_open', 'state_total_terrain_parks', and 'state_total_nightskiing_ac',
#respectively
#Finally, add a call to the reset_index() method (we recommend you experiment with and without this to see
#what it does)
neighbourhood_summary = airbnb_data.groupby('neighbourhood_cleansed').agg(
    neighborhood_cleansed_id_number=pd.NamedAgg(column='id', aggfunc='size'), #could pick any column here
    neighborhood_cleansed_review_scores_rating=pd.NamedAgg(column='review_scores_rating', aggfunc='mean'),
    neighborhood_cleansed_review_scores_location=pd.NamedAgg(column='review_scores_location', aggfunc='mean'),
    neighborhood_cleansed_bedrooms=pd.NamedAgg(column='bedrooms', aggfunc='mean'),
    neighborhood_cleansed_number_of_reviews=pd.NamedAgg(column='number_of_reviews', aggfunc='sum'),
    neighborhood_cleansed_price=pd.NamedAgg(column='price', aggfunc='mean')
).reset_index()
neighbourhood_summary.head()
```
